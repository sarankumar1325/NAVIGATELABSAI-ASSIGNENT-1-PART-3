**Import the libraires**

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

**Generate the Dataset**

```python
X = np.array([
    [1, 1],
    [1.5, 2],
    [3, 4],
    [5, 7],
    [3.5, 5],
    [4.5, 5],
    [3.5, 4.5]
])
```

**Apply the Kmeans Clustering**

```python
kmeans = KMeans(n_clusters=2, init=np.array([[1, 1], [1.5, 2]]), n_init=1, random_state=42)
kmeans.fit(X)
```

```python
cluster_assignments = kmeans.labels_
centroids = kmeans.cluster_centers_
```

**Visualsie the Cluster**

```python
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=cluster_assignments, cmap='viridis', s=100, marker='o', edgecolor='k', label='Data Points')
plt.scatter(centroids[:, 0], centroids[:, 1], s=300, c='red', marker='X', label='Centroids')
plt.title("K-means Clustering Visualization")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.grid()
plt.show()
```

**Silhouette Score**

```python
sil_score = silhouette_score(X, cluster_assignments)
print(f"Silhouette Score: {sil_score:.2f}")
```
